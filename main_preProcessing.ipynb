```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_dt=pd.read_csv('Dataset/train.csv')
test_dt=pd.read_csv('Dataset/test.csv')
```

```python
display(train_dt)
```

```python
display(test_dt)
```

```python
#pre-processing the data
```

```python
survived = train_dt['Survived']
train_dt = train_dt.drop(['Survived'],axis=1) #coloum are axis - 1 and rows are coloumn axis =0
```

```python
display(train_dt)
```

```python
train_idx=train_dt['PassengerId']
test_idx=test_dt['PassengerId']

combined_dt=pd.concat([train_dt,test_dt]).reset_index(drop=True)
display(combined_dt)
```

```python
print('Null values :' )
print(combined_dt.isnull().sum()) # provides us infromation about each coloumn and number of null values it has 
```

```python
#figuring out how to handle NaN values in all the coloumn
y=combined_dt['Age'].value_counts()
x=y.index.values
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.title('All ages')
plt.show()
```

```python
class_age= combined_dt[['Pclass', 'Age']]

for i in range(1,4):
    plt.figure(figsize=(10,5))
    plt.title('class'+ str(i))
    y=class_age[class_age['Pclass']== i]['Age'].value_counts()
    x=y.index.values
    plt.bar(x,y)
    plt.show()
```

```python
median_class = class_age.groupby('Pclass').median()
display(median_class)
```

```python
for i in range(3):
    idx=np.where((combined_dt['Pclass']==i+1) & (combined_dt['Age'].isnull()))[0]
    combined_dt.loc[idx,'Age'] =median_class.values[i][0]
```

```python
display(combined_dt)
```
